```python
import pandas as pd
import matplotlib.pyplot as plt
import shap
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import ExtraTreesClassifier
import pickle 
import numpy as np
# 将 np.bool 替换为 np.bool_
np.bool = np.bool_
```

```python
# 加载预测数据集
data = pd.read_csv('无重复值预测结果 - 副本.csv')
```

```python
import pickle# 加载训练好的ExtraTreeClassifier模型
model = ExtraTreesClassifier()
model = pickle.load(open("ET1.dat","rb"))
```

```python
# 拟合模型
X = data.drop(columns=['stablity'])
y = data['stablity']
```

```python
# 初始化 SHAP explainer
explainer = shap.Explainer(model, X)

# 计算 SHAP 值
shap_values = explainer(X)
```

```python
# 将 shap 值转换为 pandas DataFrame
shap_df = pd.DataFrame(shap_values.values[:,:,1], columns=X.columns)
```

```python

# 绘制蜂群图
shap.summary_plot(shap_values.values[:,:,1], X, show=False, plot_type='dot')
plt.tight_layout()
plt.show()
```

```python
# 绘制第一个特征 'mean radius' 的依赖关系图，对正例和反例分别进行着色
plt.rcParams.update({'font.size': 50})
shap.dependence_plot('CB', shap_values.values[:,:,1], X, 
                     interaction_index=None, alpha=1.0, color='Indigo',show=False)
plt.tight_layout()
plt.savefig('CB.png')  # 保存为 PNG 格式
plt.show()
```
